Parse data (Code by Devin Pleuler in his [handbook](https://github.com/devinpleuler/analytics-handbook))

```python
base_url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/"
comp_url = base_url + "matches/{}/{}.json"
match_url = base_url + "events/{}.json"

def parse_data(competition_id, season_id):
    matches = rq.get(url=comp_url.format(competition_id, season_id)).json()
    match_ids = [m['match_id'] for m in matches]

    all_events = []
    for match_id in tqdm(match_ids):

        events = rq.get(url=match_url.format(match_id)).json()

        shots = [x for x in events if x['type']['name'] == "Shot"]
        for s in shots:
            attributes = {
                "match_id": match_id,
                "team": s["possession_team"]["name"],
                "player": s['player']['name'],
                "x": s['location'][0],
                "y": s['location'][1],
                "outcome": s['shot']['outcome']['name'],
            }
            all_events.append(attributes)
            
    return pd.DataFrame(all_events)
```

```python
df = parse_data(43,3)
```

# Voronoi 

Trying to plot Voronoi diagram representing space domination

```python
from mplsoccer.pitch import Pitch
from pprint import pprint
```

```python
base_url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/"
comp_url = base_url + "matches/{}/{}.json"
match_url = base_url + "events/{}.json"

def parse_all_data(competition_id, season_id):
    
    matches = rq.get(url=comp_url.format(competition_id, season_id)).json()
    
    match_id = random.choice([m['match_id'] for m in matches])
    match_round = [m['competition_stage']['name'] for m in matches if m['match_id'] == match_id][0]
    all_events = []
    url = match_url.format(match_id)
    events = rq.get(url).json()
    
    
    lineup_events = [x for x in events if x['type']['name']=='Starting XI']
    lineups = []
    teams = []
    for x in lineup_events:
        teams.append(x['team']['name'])
        lineups.append(x['tactics']['lineup'])
    
    for x,l in enumerate(lineups):
        for p in l:
            attributes = {
                        "match_id": match_id,
                        "team" : teams[x],
                        "player": p['player']['name'],
                        "pos_id": p['position']['id'],
                        "pos_name": p['position']['name'],
                        "number": p['jersey_number']
                    }
            all_events.append(attributes)

    return pd.DataFrame(all_events),match_round
```

```python
df_matchs,match_round = parse_all_data(43,3)
df_matchs.head()
```
